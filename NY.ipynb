```python
import pandas as pd
df = pd.read_csv('NY.csv')
df
```

```python
newdf = df.drop(columns = ['BROKERTITLE','ADDRESS','STATE','MAIN_ADDRESS','ADMINISTRATIVE_AREA_LEVEL_2','STREET_NAME','LONG_NAME','FORMATTED_ADDRESS','LATITUDE','LONGITUDE'])
newdf
```

```python
newdf1 = pd.get_dummies(newdf,dtype=int)
newdf1
```

```python
x = newdf1.drop(columns =['PRICE'])
y = newdf1['PRICE']
x
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
md = Lasso(alpha=50,max_iter=100,tol=0.1)
md.fit(x_train,y_train)
```

```python
md.score(x_train,y_train)
```

```python
md.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
md1 = Ridge(alpha=50,max_iter=100,tol=0.1)
md1.fit(x_train,y_train)
```

```python
md1.score(x_train,y_train)
```

```python
md1.score(x_test,y_test)
```
